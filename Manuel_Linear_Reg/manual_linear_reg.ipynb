```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
import sklearn.linear_model
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1) #y = 4 + 3x + noise
```

```python
plt.plot(X,y,"b.")
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)), X]
```

![](gorsel.png)

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
# y = 4 + 3x + noise
# y = 4.59 + 2.95x
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
```

```python
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X,y,"b.")
plt.show()
```
